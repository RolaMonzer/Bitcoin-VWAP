L’objectif de ce projet est de calculer les Volume Weighted Average Price (VWAP) horaires du Bitcoin, pour chacun des exchanges à disposition dans le dossier "data".

On commence par importer les libraries qu'on va utiliser lors de notre projet (pandas et glob).

```python
import pandas as pd
import glob
```

Après, on veux obtenir les noms des fichiers qui existe dans le dossier "data" et les stocker dans une liste "all_files". On a choisi de faire cela pour faire une boucle sur les dossiers au lieux d'executer les mêmes commandes pour chacun.

On a choisi d'utiliser la fonction glob qui nous rend tout les fichier csv dans notre dossier. 

```python
path = 'data' 
all_files = glob.glob(path + "/*.csv")

all_files
```

Maintenant, que l'array "all_files" contient les noms du fichiers, on peut faire une boucle qui lit et manipule les données et calcul le vwap. En gros, les 3 parties du projet se situe dans cette boucle.

On commence par initialiser 2 data frames vides "DF" et "Global", "DF" sera le data frame final de cette partie où on affiches les valeurs VWAP par heure pour chaque exchange et "Global" sera le résulat d'une concatenation de tous les trades sans spécifier les exchanges. 

```python
DF =  pd.DataFrame()
Global = pd.DataFrame()

for filename in all_files:
    
    # Read Input
    
    df = pd.read_csv(filename) # lire chaque fichier qui existe dans la liste all_files
    Global = pd.concat([Global,df]) # obtimiser la boucle pour n'utiliser pas une autre dans l'étape du calcul global
    filename = filename[5:9] # renommer la variable filename pour avoir le nom de l'exchange
    
    # Manipulation de données
    
    df["timestamp"] = pd.to_datetime(df["timestamp"]) # changer le type de la colonne "timestamp" pour un datetime
    df.set_index("timestamp", inplace=True) # indexer le data frame par "timestamp"
    df["price*amount"] = df["price"]*df["amount"] # calculer le prix*volume pour chaque trade exchange
    df= df.resample("1H").sum() # regrouper les données par heure et sommer sur les autres variables 
                                # pour obtenir sum(amount) et sum(prix*amount) 
    
    # Calcul de VWAP
    
    df['vwap_'+filename] = df["price*amount"]/df["amount"] # ajouter un colonne avec les valeurs de vwap et le nom de l'exchange
    #df.fillna(0)  # normalment, on traite les NaN à ce stade, mais on a laissé cette étape pour après l'ajout des vwap global
    df= df['vwap_'+filename] # récuperer seul la colonne VWAP de chaque exchange 
    DF = pd.concat([DF,df], axis=1) # concatener cette colonne avec le data frame final de tous les VWAP de chaque exchange.

DF # aficher le data frame indexer par l'heure ayant 11 colonnes chacune avec les valeurs VWAP par trade
```

Maintenant, notre but est d'ajouter les valeurs de vwap global ayant un data frame qui contient tous les trade avec leurs prix et volume (sans un index sur les trades). On fait une manipulation de données et un calcul pareil que pour 1 exchange. 

```python
# Manipulation de données
Global["timestamp"] = pd.to_datetime(Global["timestamp"])
Global.set_index("timestamp", inplace=True)
Global["price*amount"] = Global["price"]*Global["amount"]
Global= Global.resample("1H").sum()
    
# Calcul de VWAP
Global['vwap_Global'] = Global["price*amount"]/Global["amount"]
Global= Global['vwap_Global']
Global
```

Finalement, on va concatener le data frame "Global" avec le data frame ayant les autres valeurs VWAP. Et on va remplacer les valeurs NaN par zéro, comme ces valeurs représente que aucune trade ne s'est passé à cette heure. En plus, on export le data frame a un fichier csv.

```python
DF = pd.concat([DF,Global],axis=1)
DF = DF.fillna(0) # changer les valeurs NaN par zéro 
DF.to_csv("./VWAP_all_trades.csv", index = True)
DF
```

Avec ce data frame on a achevé notre but de calculer les differents vwap par exchange et par heure, et les vwap global par heure.

# Aller plus loin

Dans cette partie extra, nous avons d'afficher les vwap dans differents graphs. Alors on commence par importer la librairie matplotlib

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

On va montrer les variations des vwap par echange en utilisant des lines plots.

```python
fig, ax = plt.subplots()

# set colormap
#note cette colormap sera repetitive si on choisi 11 coleurs, alors on choisi 6 de celle la et 5 couleurs d'une autre
clrs = plt.get_cmap('Dark2', 7)
i=0 # l'index pour faire l'iteration des couleurs

# plot line plots pour les vwap de chaque exchange
for filename in all_files:
    i = i+1
    filename = filename[5:9]
    ax.plot_date(DF.index, DF['vwap_'+filename], xdate=True, ydate=False, ls='-', color=clrs(i), marker='', label=filename)
    if i == 6: 
        # re-initialiser la colormap quand on a utiliser ces couleurs
        clrs = plt.get_cmap('Set1', 6)
        i=0
        
# plot line plot pour les vwap global        
ax.plot_date(DF.index, DF['vwap_Global'], xdate=True, ydate=False, ls='-', color='k', marker='', label='Global')  

# changer le format des dates pour contenir seulement l'heure
ax.xaxis.set_major_formatter(mdates.DateFormatter('%H'))

# ajouter une legende, un titre, et des labels sur les axes
ax.legend(loc='center left', bbox_to_anchor= (1.04, 0.5), ncol=1, borderaxespad=0, frameon=False)
ax.set_title('VWAP horaires du Bitcoin', loc= 'center', fontsize=16)
ax.set_xlabel('heure', fontsize=15)
ax.set_ylabel('vwap', fontsize=15)

# ajouter une gird et enlever le carde de la figure
ax.grid(b=True, which='major')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

# on a trouver que la plupart des valeurs (tous sauf NaN) respecte cette limite
# donc c'est la partie du graphe interessante, qu'on veut mieux visaliser
ax.set_ylim(47000,52000)

plt.show()
```

Ce graphe nous donne pas trop d'informations sur les differents exchanges, d'ou la nécessiter d'autre graphics pour bien visualiser les échanges de vwap qu'on va laisser pour ceux qui sont intéresser.
